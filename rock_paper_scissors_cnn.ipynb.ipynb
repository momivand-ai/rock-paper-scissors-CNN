```python
# TODO: تابع نمایش تصاویر رندوم را کامل کنید.


def visualize_two_sample_images(directory, title="Sample Images"):
    """
    دو تصویر تصادفی از مسیر مشخص‌ شده بارگذاری و نمایش می‌دهد.
    """
    # گرفتن لیست تمام فایل‌های موجود در پوشه
    all_images = os.listdir(directory)

    # انتخاب ۲ تصویر به صورت کاملاً تصادفی
    random_images = random.sample(all_images, 2)

    # ایجاد یک قاب برای نمایش تصاویر (۱ سطر و ۲ ستون)
    plt.figure(figsize=(10, 4))

    for i, img_name in enumerate(random_images):
        # ساخت مسیر کامل فایل تصویر
        img_path = os.path.join(directory, img_name)

        # خواندن تصویر
        img = mpimg.imread(img_path)

        # قرار دادن تصویر در بخش مربوطه
        plt.subplot(1, 2, i + 1)
        plt.imshow(img)
        plt.title(f"{title} (Sample {i+1})")
        plt.axis('off') # پنهان کردن محورهای مختصات برای زیبایی بیشتر

    plt.show()

# فراخوانی تابع برای هر سه کلاس
visualize_two_sample_images(PAPER_DIR, title="Paper")
visualize_two_sample_images(ROCK_DIR, title="Rock")
visualize_two_sample_images(SCISSORS_DIR, title="Scissors")
```

        از تابع image_dataset_from_directory استفاده کنید تا تصاویر به‌صورت خودکار به تانسورهای قابل استفاده در مدل تبدیل شوند.

* [Load & Preprocess Images | Tensorflow](https://www.tensorflow.org/tutorials/load_data/images)
* [image_dataset_from_directory | Tensorflow](https://www.tensorflow.org/api_docs/python/tf/keras/preprocessing/image_dataset_from_directory)
* [How to Use](https://datascience.stackexchange.com/questions/87469/how-to-load-all-images-using-image-dataset-from-directory-function)

<br><br>

```python


# مشخص کردن اندازه تصاویر و اندازه هر بچ
IMG_SIZE = (150, 150)  # تصاویر به این اندازه (۱۵۰ در ۱۵۰ پیکسل) تغییر ابعاد می‌دهند
BATCH_SIZE = 32        # در هر مرحله ۳۲ تصویر به مدل داده می‌شود

# بارگذاری داده‌های آموزش (۸۰ درصد داده‌ها)
train_dataset = tf.keras.utils.image_dataset_from_directory(
    BASE_DIR,
    validation_split=0.2,  # ۲۰ درصد داده‌ها را کنار بگذار
    subset="training",     # از ۸۰ درصد باقی‌مانده برای "آموزش" استفاده کن
    seed=123,              # یک عدد ثابت برای تصادفی‌سازی یکسان
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE
)

# بارگذاری داده‌های اعتبارسنجی (۲۰ درصد داده‌ها)
validation_dataset = tf.keras.utils.image_dataset_from_directory(
    BASE_DIR,
    validation_split=0.2,  # همان ۲۰ درصد را در نظر بگیر
    subset="validation",   # این بار بخش "اعتبارسنجی" را جدا کن
    seed=123,              # این عدد باید دقیقاً مثل بالا باشد
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE
)
```

<div dir='rtl'>
  <font size=4>
    <li>
        در این مرحله باید داده‌های آموزش و اعتبارسنجی را برای آموزش بهینه مدل آماده‌سازی کنید.
        برای این کار:
        <li>        از .cache() برای ذخیره‌سازی داده‌ها در حافظه استفاده کنید تا خواندن آن‌ها در هر epoch سریع‌تر انجام شود.</li>
        <li>        از .shuffle() برای ایجاد تنوع در ترتیب داده‌های آموزشی استفاده کنید تا مدل دچار overfitting نشود.</li>
        <li>        از .prefetch() برای بارگذاری داده‌ها به‌صورت غیرهمزمان استفاده کنید تا بهره‌وری پردازنده/کارت گرافیک افزایش یابد.</li>
    </li>
  </font>
</div>
<br><br>

```python
# TODO : کش کردن داده‌ها در حافظه برای افزایش سرعت آموزش (فقط در صورت کافی بودن حافظه RAM)

# تنظیم خودکار حافظه پنهان (Buffer) توسط خود تنسورفلو
AUTOTUNE = tf.data.AUTOTUNE

# ۱. بهینه‌سازی دیتاست آموزش
train_dataset = train_dataset.cache().shuffle(buffer_size=1000).prefetch(buffer_size=AUTOTUNE)

# ۲. بهینه‌سازی دیتاست اعتبارسنجی
# نکته: داده‌های اعتبارسنجی نیازی به shuffle (قاطی شدن) ندارند چون فقط برای امتحان گرفتن هستند.
validation_dataset = validation_dataset.cache().prefetch(buffer_size=AUTOTUNE)

print("✅ داده‌های آموزش و اعتبارسنجی با موفقیت برای پردازش سریع بهینه‌سازی شدند.")
```

```python

def visualize_augmented_images(dataset, augmentation_layer, num_samples=9):
    """
    نمایش تعدادی تصویر افزایش‌ یافته (Augmented) از یک دیتاست آموزشی.
    """
    # ۱. گرفتن فقط یک بچ (دسته) از داده‌ها
    for images, labels in dataset.take(1):
        break

    # کنترل خطا: تعداد درخواستی نباید بیشتر از حجم یک بچ باشد
    if num_samples > images.shape[0]:
        num_samples = images.shape[0]
        print(f"⚠️ تعداد درخواستی بیشتر از حجم بچ بود؛ به {num_samples} تغییر یافت.")

    # ۲. اعمال لایه افزایش داده روی تمام تصاویر بچ به صورت یکجا (پر سرعت)
    augmented_images = augmentation_layer(images, training=True)

    # ۳. محاسبه هوشمند تعداد سطر و ستون‌ها بر اساس تعداد تصاویر درخواستی
    cols = int(math.ceil(math.sqrt(num_samples)))
    rows = int(math.ceil(num_samples / cols))

    # تنظیم اندازه کادر کلی نمودار
    plt.figure(figsize=(cols * 3, rows * 3))

    # ۴. رسم تصاویر در قالب یک شبکه تصویری مرتب
    for i in range(num_samples):
        plt.subplot(rows, cols, i + 1)

        # تبدیل به فرمت تصویری استاندارد پایتون (uint8) و نمایش آن
        plt.imshow(augmented_images[i].numpy().astype("uint8"))
        plt.title(f"Sample {i+1}")
        plt.axis("off") # حذف خطوط مدرج کنار عکس

    plt.tight_layout() # تنظیم فاصله‌ها برای جلوگیری از روی هم افتادن متن‌ها
    plt.show()
```

  <font size=4>
    <li>
        این بخش از باید پیش‌پردازش (Preprocessing) روی داده‌های تصویری قبل از ورود به مدل انجام دهید.
         در اینجا هدف اصلی نرمال‌سازی مقادیر پیکسل‌ها است، به این صورت که مقادیر اصلی که در بازه [0, 255] هستند، به بازه [0.0, 1.0] تبدیل می‌شوند.
         یک تابع تعریف شده که هر تصویر را از نوع int به float تبدیل می‌کند و سپس آن را بر ۲۵۵ تقسیم می‌کند تا نرمال شود. این تابع را کامل کنید و سپس  روی کل داده‌های آموزش و اعتبارسنجی با استفاده از متد map اعمال شده است.
    </li>
  </font>
<br><br>

```python
# تعریف تابع پیش‌پردازش شامل نرمال‌سازی مقادیر پیکسل


def preprocess(image, label):
    # تبدیل نوع داده تصویر از int به float32 برای محاسبات دقیق اعشاری
    image = tf.cast(image, tf.float32)

    # تقسیم بر 255 برای تبدیل بازه [0, 255] به [0.0, 1.0]
    image = image / 255.0

    return image, label

# ۲. اعمال تابع روی کل داده‌های آموزش و اعتبارسنجی با استفاده از متد map
# پارامتر AUTOTUNE کمک می‌کند این کار به صورت موازی و با حداکثر سرعت انجام شود
train_dataset = train_dataset.map(preprocess, num_parallel_calls=tf.data.AUTOTUNE)
validation_dataset = validation_dataset.map(preprocess, num_parallel_calls=tf.data.AUTOTUNE)

print("✅ تابع پیش‌پردازش و نرمال‌سازی با موفقیت روی تمام داده‌ها اعمال شد.")
```

        در این مرحله،  باید یک مدل شبکه عصبی کانولوشنی (CNN) طراحی کنید که شامل لایه‌های استخراج ویژگی، کاهش ابعاد و لایه‌های تمام‌متصل برای طبقه‌بندی تصاویر به سه دسته مختلف باشد. همچنین، مدل باید کامپایل شده و آماده آموزش باشد.

```python
# TODO: ساخت مدل شبکه عصبی کانولوشنی (CNN) برای طبقه‌بندی تصاویر

# ۱. طراحی معماری شبکه عصبی کانولوشنی (CNN)
model = models.Sequential([
    # تعیین ابعاد ورودی: تصاویر ۱۵۰ در ۱۵۰ پیکسل با ۳ کانال رنگی (RGB)
    layers.Input(shape=(150, 150, 3)),

    # بلوک اول کانولوشن و کاهش ابعاد
    layers.Conv2D(32, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    # بلوک دوم کانولوشن و کاهش ابعاد
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    # بلوک سوم کانولوشن و کاهش ابعاد
    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    # بلوک چهارم کانولوشن و کاهش ابعاد (برای استخراج الگوهای عمیق‌تر و پیچیده‌تر دست)
    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    # تبدیل نقشه ویژگی‌های ۲ بعدی به یک بردار ۱ بعدی تخت
    layers.Flatten(),

    # لایه تمام‌متصل (Dense) اصلی با ۵۱۲ نورون برای تصمیم‌گیری
    layers.Dense(512, activation='relu'),

    # لایه دراپ‌اوت (Dropout) برای حذف تصادفی ۵۰٪ نورون‌ها جهت جلوگیری از حفظ کردن طوطی‌وار داده‌ها (Overfitting)
    layers.Dropout(0.5),

    # لایه خروجی با ۳ نورون (برای ۳ کلاس سنگ، کاغذ و قیچی)
    # تابع فعال‌ساز softmax خروجی را به صورت درصد احتمال بین کلاس‌ها تقسیم می‌کند
    layers.Dense(3, activation='softmax')
])

# ۲. کامپایل کردن مدل (تنظیم استراتژی آموزش)
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# نمایش کادر خلاصه معماری مدل
print("=" * 60)
print("🧠 خلاصه معماری لایه‌های شبکه عصبی طراحی‌شده:")
print("=" * 60)
model.summary()
```

```python
def plot_training_history(history):
    """
    رسم گراف‌های دقت و خطا برای داده‌های آموزش و اعتبارسنجی.
    """
    # ۱. استخراج اطلاعات دقت و خطا از شیء history
    acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']

    loss = history.history['loss']
    val_loss = history.history['val_loss']

    # مشخص کردن تعداد اپوک‌ها برای محور افقی
    epochs_range = range(1, len(acc) + 1)

    # ۲. تنظیم ابعاد پنجره کلی نمودارها (یک سطر و دو ستون)
    plt.figure(figsize=(14, 5))

    # --- نمودار اول: دقت (Accuracy) ---
    plt.subplot(1, 2, 1)
    plt.plot(epochs_range, acc, label='Training Accuracy', color='#1f77b4', linewidth=2, marker='o')
    plt.plot(epochs_range, val_acc, label='Validation Accuracy', color='#ff7f0e', linewidth=2, marker='x')
    plt.title('Training and Validation Accuracy', fontsize=12, fontweight='bold', pad=10)
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend(loc='lower right')
    plt.grid(True, linestyle='--', alpha=0.6)

    # --- نمودار دوم: خطا (Loss) ---
    plt.subplot(1, 2, 2)
    plt.plot(epochs_range, loss, label='Training Loss', color='#d62728', linewidth=2, marker='o')
    plt.plot(epochs_range, val_loss, label='Validation Loss', color='#2ca02c', linewidth=2, marker='x')
    plt.title('Training and Validation Loss', fontsize=12, fontweight='bold', pad=10)
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend(loc='upper right')
    plt.grid(True, linestyle='--', alpha=0.6)

    # تنظیم فاصله‌ها برای زیباتر شدن خروجی
    plt.tight_layout()
    plt.show()

# ۳. فراخوانی تابع برای نمایش نمودارها
plot_training_history(history)
```

```python
import numpy as np

def predict_image(model, image_path, class_names, img_size=(150, 150)):
    """
    بارگذاری و پیش‌پردازش یک تصویر و پیش‌بینی کلاس آن.
    """
    # ۱. بارگذاری تصویر و تغییر اندازه آن به ابعاد مورد نیاز مدل
    img = tf.keras.utils.load_img(image_path, target_size=img_size)

    # ۲. تبدیل تصویر بارگذاری‌شده به یک آرایه عددی (numpy array)
    img_array = tf.keras.utils.img_to_array(img)

    # ۳. اعمال پیش‌پردازش: تبدیل به float32 و نرمال‌سازی بین [0, 1] (دقیقاً مثل داده‌های آموزش)
    img_array = tf.cast(img_array, tf.float32) / 255.0

    # ۴. اضافه کردن بعد بچ (Batch Dimension)؛ تبدیل شکل از (150, 150, 3) به (1, 150, 150, 3)
    img_array = tf.expand_dims(img_array, 0)

    # ۵. دریافت پیش‌بینی از مدل
    predictions = model.predict(img_array)

    # ۶. پیدا کردن ایندکس کلاسی که بیشترین احتمال را دارد و محاسبه درصد اطمینان
    predicted_class_idx = np.argmax(predictions[0])
    predicted_class = class_names[predicted_class_idx]
    confidence = np.max(predictions[0]) * 100

    # ۷. نمایش تصویر ورودی به همراه عنوان شامل نام کلاس و درصد اطمینان
    plt.figure(figsize=(5, 5))
    plt.imshow(img)
    plt.title(f"Prediction: {predicted_class}\nConfidence: {confidence:.2f}%", fontsize=12, fontweight='bold')
    plt.axis("off")
    plt.show()

    # بازگرداندن خروجی‌ها
    return predicted_class, confidence
```

```python
# تعریف نام کلاس‌ها به ترتیب حروف الفبا یا ترتیب پوشه‌ها در دیتاست شما
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

# ۱. تعریف نام کلاس‌ها (به همان ترتیبی که دیتاست باز می‌کند)
my_class_names = ['paper', 'rock', 'scissors']

# ۲. استخراج یک بچ از داده‌های اعتبارسنجی و برداشتن اولین عکس و برچسب واقعی آن
for test_images, test_labels in validation_dataset.take(1):
    sample_image = test_images[0]
    true_label_idx = test_labels[0].numpy()
    break

# ۳. آماده‌سازی تصویر برای مدل (افزودن بعد بچ)
# نکته: این عکس از قبل توسط تابع preprocess نرمال‌سازی شده و آماده است
img_array = tf.expand_dims(sample_image, 0)

# ۴. انجام پیش‌بینی توسط مدل
predictions = model.predict(img_array)
predicted_class_idx = np.argmax(predictions[0])

# ۵. استخراج نام کلاس‌ها و محاسبه درصد اطمینان
predicted_label = my_class_names[predicted_class_idx]
true_label = my_class_names[true_label_idx]
confidence = np.max(predictions[0]) * 100

# ۶. نمایش تصویر و خروجی با رنگ‌بندی هوشمند برای گیت‌هاب
plt.figure(figsize=(5, 5))
# چون مقادیر پیکسل‌ها بین 0.0 و 1.0 (float32) هستند، matplotlib به زیبایی آن را رسم می‌کند
plt.imshow(sample_image.numpy())

# اگر پیش‌بینی درست بود رنگ سبز و اگر غلط بود رنگ قرمز نشان داده شود
title_color = 'green' if predicted_label == true_label else 'red'

plt.title(
    f"True Label: {true_label}\n"
    f"Predicted: {predicted_label} ({confidence:.2f}%)",
    fontsize=12,
    fontweight='bold',
    color=title_color,
    pad=10
)
plt.axis("off")
plt.show()
```
